คำตอบที่ถูกต้องที่สุด

The most correct answer

````python
def sorting(input_number,type=True):
  n = len(input_number)
  if type == True:
    for i in range(n):
      for j in range(0,n-i-1):
        if input_number[j] > input_number[j+1]:
          input_number[j], input_number[j+1] = input_number[j+1], input_number[j]
  else:
    for i in range(n):
      for j in range(0,n-i-1):
        if input_number[j+1] > input_number[j]:
          input_number[j], input_number[j+1] = input_number[j+1],input_number[j]
  return input_number
````

````python
def analyze_letters(list_aa, word, list_name):
  # เก็บตัวอักษรใน list_aa ว่ามีตัวอะไรบ้าง
    word_split = []
    for i in list_aa:
        for j in i: # ตัวอักษรในคำที่ศัพท์ที่มี
            if j not in word_split:
                word_split.append(j)

   #หาตัวอักษรที่ต้องการ
    word_collect = []
    for w in word:
        letter_collect = []
        for l in w:
            if l in word_split:
                letter_collect.append(l)
            # else:
            #     print(f'The letter {l} not in list_aa')
        word_collect.append(''.join(letter_collect)) #รวมตัวอักษรในคำที่ต้องการหา

    print(f'List of word found --> {word_collect}')
    for c in word: # print คำที่ต้องการหาก่อน
        print(c)
        for l in c:
            found = False #Tracks Status, Avoids Redundant Prints, Ensures Complete Check
            for i in list_aa:
                if l in i:
                    print(f'The letter {l} can be found in the word {i} in {list_name}')
                    found = True
                    break # ถ้าเจอใน 1 คำของ list_aa แล้วออกจากลูป  #break statement in the inner loop to exit once the letter is found in list_aa.
            if not found:
              print(f'The letter {l} not in {list_name}')
        print('-'*100)
````
