```python
# 1 record → ["Object=...", "Outcome=..."]
transactions = dataupdate.apply(
    lambda r: [f"Object={r['Object of search']}",
               f"Outcome={r['Outcome']}"],
    axis=1
).tolist()
```

```python
min_sup = 0.065
frequentItemsets = list(apriori(transactions, min_support=min_sup))

for i in frequentItemsets:
    if len(i[0]) > 1:   # ถ้ามีมากกว่า 1 item
        print("yes")
    print(i)
```

support = 0.11096 (~11.1%)
~11.1% ของเหตุการณ์ทั้งหมดเป็นคู่ (Article for use in theft ∧ NFA)

กฎ 1 (ฐานว่าง → คู่) – ข้ามได้ (เชิงเทคนิค)

confidence = support, lift = 1.0

กฎ 2: {Object=Article for use in theft} → {Outcome=NFA}

confidence = 0.64516 (64.5%) = ถ้าเจอ Article for use in theft แล้ว 64.5% จบ NFA

lift = 0.91567 (< 1) = โอกาส NFA ต่ำกว่า*ค่าเฉลี่ยทั้งชุดเล็กน้อย
(baseline NFA ≈ 0.64516 / 0.91567 ≈ 70.5%)

กฎ 3: {Outcome=NFA} → {Object=Article for use in theft}

confidence = 0.15748 (15.7%) = ในเคสที่จบ NFA มี ~15.7% ที่เป็น Article for use in theft

```python
def extract_rules(records):
    rows = []
    for r in records:
        for stat in r.ordered_statistics:
            A = list(stat.items_base)
            C = list(stat.items_add)
            if len(C)==1 and C[0].startswith('Outcome=') and any(x.startswith('Object=') for x in A):
                rows.append({
                    'antecedent': ', '.join(A),
                    'consequent': C[0],
                    'support': r.support,
                    'confidence': stat.confidence,
                    'lift': stat.lift
                })
    import pandas as pd
    return (pd.DataFrame(rows)
              .sort_values(['lift','confidence','support'], ascending=False)
              .reset_index(drop=True))

rules_df = extract_rules(frequentItemsets)
rules_df
```
